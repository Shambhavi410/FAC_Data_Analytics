```python
import pandas as pd
```

```python
# Task 2.1: Create a DataFrame from the data of students above
data={
    "Roll No.":[2301,2310,2302,2304,2306,2305,2307,2312,2315,2316,2320,2308,2309,2314,2319],
    "Name":["Alice","Smith","Bob","Charlie","Eric","Diana","John","Joey","Bean","Charlie","Jack","Jill","George","Tabitha","Phoebe"],
    "Age":[24,None,22,23,None,24,None,30,26,23,22,21,24,31,24],
    "Marks":[85,None,78,92,95,88,None,64,72,92,83,69,89,100,91]
}
data=pd.DataFrame(data)
print(data)
```

***Hint:*** **NaN value** is represented by **None**

```python
# Task 2.2: Inspect the DataFrame(use the most basic functions to know the statistical quantities)
print(data[['Age','Marks']].describe())
```

```python
# Task 2.3: Select the Marks column and Print it
print(data['Marks'])
```

```python
# Task 2.4: Access the second row
second_row=data.iloc[1]
print(second_row)
```

```python
# (BONUS) Task 2.5: Access only the rows where Marks > 80
x=data[data['Marks']>80]
print(x)
```

```python
# (BONUS, but really fun try to find it do it how :) ) Task 2.6: Make a function to assign grade
# (i) A if marks > 85
# (ii) B if marks > 75
# (iii) C if marks > 65
def grade(data):
    data['Grade']=None
    for i in range (data.shape[0]):
        if data['Marks'][i]>85:
            data.loc[i,'Grade']='A'
        elif data['Marks'][i]>75 and data['Marks'][i]<=85:
            data.loc[i,'Grade']='B'
        elif data['Marks'][i]>65 and data['Marks'][i]<=75:
            data.loc[i,'Grade']='C'
        else:
          data.loc[i,'Grade']='No Grade'
grade(data)
print(data)
```

```python
# Task 2.7: Rename Marks column to Score
data.rename(columns={"Marks": "Score"}, inplace=True)
data
```

```python
# Task 2.8: Replace NaN in Age with mode and NaN in Marks with mean
for i in range (data.shape[0]):
    if pd.isnull(data['Age'][i]):
        data.loc[i,'Age']=data['Age'].mode()[0]
    if pd.isnull(data['Score'][i]):
        data.loc[i,'Score']=data['Score'].mean()
data
```

```python
# Task 2.9: Remove duplicates
data=data.drop_duplicates(subset='Roll No.')
data
```

```python
# Task 2.10: Group the data by Grade and print the data
group= data.groupby('Grade')
for i, j in group:
    print(f"Grade: {i}")
    print(j)
    print("\n")
```

```python
# Task 2.11: Sort Data in increasing order of Age
data.sort_values(by='Age')
```

```python
# Task 2.12: Export the dataframe to CSV
data.to_csv('students.csv')
```

# ***Part 3: Matplotlib***

```python
import matplotlib.pyplot as plt
```

```python
#Applying the grade function again as the data has been edited by replacing NaN values with mean
data.rename(columns={"Score": "Marks"}, inplace=True)
grade(data)
# Task 3.1: Plot a Pie Chart depicting the percentage of grades that the students have got.
x=data['Grade'].value_counts()
label=x.index
perc=[]
for i in x:
  perc.append(i/data['Grade'].count()*100)
plt.pie(perc, labels=label, autopct='%1.1f%%')
plt.title('Percentage of Grades')
plt.axis('equal')
plt.show()
```

```python
# Task 3.2: Plot a Line Graph of Roll No. vs Marks
plt.plot(data.sort_values('Roll No.')["Roll No."], data["Marks"],marker='o',label='Marks')
plt.xlabel('Roll No.')
plt.ylabel('Marks')
plt.title('Roll No. vs Marks')
plt.legend()
```

```python
# Task 3.3: Plot a Histogram of for showing the Age distribution
plt.hist(data['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution')
```
