**Q14: Identify columns with missing values.**

**Decide how to handle missing values:**

  **- Drop rows/columns with excessive missing values.***

  **- Impute missing values for `hospital_beds_per_thousand` and `gdp_per_capita` using the median or mean.**

```python
countries_df['hospital_beds_per_thousand']=countries_df['hospital_beds_per_thousand'].fillna(countries_df['hospital_beds_per_thousand'].mean())
countries_df['gdp_per_capita']=countries_df['gdp_per_capita'].fillna(countries_df['gdp_per_capita'].mean())
countries_df['gdp']=countries_df['population']*countries_df['gdp_per_capita']
countries_df
```

**Q15: Encode the `continent` column using one-hot encoding.**

**- Check the updated dataset to ensure encoding was applied correctly.**

```python
x=countries_df['continent'].unique()
countries_df[[x]]=0
for i in range(countries_df.shape[0]):
  for j in x:
    if countries_df.loc[i,'continent'] == j:
      countries_df.loc[i,j] = 1
countries_df.drop(columns='continent',inplace=True)
countries_df
```

**Q16: Scale the numerical features (`population`, `life_expectancy`, `hospital_beds_per_thousand`, `gdp_per_capita`) using standardization or normalization.**

**Display the summary statistics of the scaled features.**

```python
import pandas as pd
from scipy import stats
countries_df['life_expectancy'] = countries_df['life_expectancy'].fillna(countries_df['life_expectancy'].mean())
scaled_data = {}
scaled_data['population'], _ = stats.boxcox(countries_df['population'])
scaled_data['life_expectancy'], _ = stats.boxcox(countries_df['life_expectancy'])
scaled_data['hospital_beds_per_thousand'], _ = stats.boxcox(countries_df['hospital_beds_per_thousand'])
scaled_data['gdp_per_capita'], _ = stats.boxcox(countries_df['gdp_per_capita'])
scaled_df = pd.DataFrame(scaled_data)
summary_statistics = scaled_df.describe()
print(summary_statistics)
```

**Q17: Create the following visualizations:**

  **- A bar plot showing the number of countries in each continent.**

  **- A scatter plot of `gdp_per_capita` vs. `life_expectancy`.**

  **- A histogram of `population` (log-scaled if necessary).**

 **- A heatmap to visualize correlations between numerical features.**

```python
import numpy as np
import matplotlib.pyplot as plt
countries_df=pd.read_csv('countries.csv')
countries_df['continent'].value_counts().plot(kind='bar')
plt.title('Number of countries in each continent')
plt.xlabel('Continent')
plt.ylabel('Number of countries')
plt.show()
plt.scatter(countries_df['gdp_per_capita'], countries_df['life_expectancy'])
plt.title('GDP per capita vs life expectancy')
plt.xlabel('GDP per capita')
plt.ylabel('Life expectancy')
plt.show()
plt.hist(countries_df['population'], bins=30)
plt.yscale('log')
plt.title('Population Distribution')
plt.xlabel('Population')
plt.ylabel('Frequency (log scaled)')
plt.show()
a=['population', 'life_expectancy', 'hospital_beds_per_thousand', 'gdp_per_capita']
correlation_matrix = countries_df[a].corr()
plt.matshow(correlation_matrix, fignum=1, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(a)), a, rotation=90)
plt.yticks(range(len(a)), a)
plt.title('Correlation heatmap of numerical features')
plt.show()
```
